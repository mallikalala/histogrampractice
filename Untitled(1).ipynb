```python
from astroquery.ipac.irsa import Irsa
Irsa.list_catalogs()   
import matplotlib.pyplot as plt
```

```python
from astroquery.ipac.irsa import Irsa
from astropy.coordinates import SkyCoord
coord = SkyCoord(121.1743, -21.5733, unit='deg', frame='galactic')
table = Irsa.query_region(coordinates=coord,
                          catalog='fp_psc', radius='0d2m0s')
print(table)
```

```python
import numpy as np
```

```python

data1 = table['ra']
data2 = table['dec']
 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
 
axes[0].hist(data1, bins=30, color='Yellow', edgecolor='black')
axes[0].set_title('Histogram 1')
 
axes[1].hist(data2, bins=30, color='Pink', edgecolor='black')
axes[1].set_title('Histogram 2')
 
plt.tight_layout()
 
plt.show()
```
